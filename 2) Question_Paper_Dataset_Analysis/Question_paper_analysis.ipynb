# Analysis on Question Paper Analysis

# Dataset Description:
This dataset contains 86 student records, detailing their scores in different quiz assessments. It includes 7 columns, where Q1 to Q6 represent scores in six different quizzes, and the Total column shows the sum of all quiz scores for each student.

Each row represents a student's performance across these quizzes. Since all values are numerical, the dataset is well-structured for analysis, such as identifying top performers, analyzing score distribution, and comparing quiz difficulty levels.

# Overview of the dataset :
The Class Marks Dataset contains records of students' performance across six quizzes (Q1 to Q6) along with a Total score, which represents the sum of all quiz marks for each student. This dataset helps in analyzing students' academic performance, identifying trends in scores, and comparing quiz difficulty levels.

By examining the dataset, we can determine average scores, top performers, weakest areas, and overall class performance. It can also be useful for teachers to assess student progress and make data-driven decisions for improvement.

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv(r"C:\Users\modil\Downloads\class_marks.csv")
df
```

```python
df.shape
```

```python
df["Q1"] = df["Q1aM4"] + df["Q1bM6"]
df["Q2"] = df["Q2aM6"] + df["Q2bM4"]
df["Q3"] = df["Q3aM5"] + df["Q3bM5"]
df["Q4"] = df["Q4aM3"] + df["Q4bM7"]
df["Q5"] = df["Q5M10"]
df["Q6"] = df["Q6aM4"] + df["Q6bM6"]
df.drop(["Q1aM4", "Q1bM6","Q2aM6","Q2bM4" ,"Q3aM5" , "Q3bM5", "Q4aM3" ,"Q4bM7","Q5M10","Q6aM4","Q6bM6"], axis=1, inplace=True) 
df
```

```python
df.fillna(0)
```

```python
data_int = df.fillna(0).astype(int)

output_file_path = (r"C:\Users\modil\Downloads\class_marks.csv")
data_int.to_csv(output_file_path, index=False)
data_int.head()
```
