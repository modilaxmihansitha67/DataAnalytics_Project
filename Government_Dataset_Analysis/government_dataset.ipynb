# Dataset Description:

The Crime Review for October 2023 dataset contains records of various crimes reported during the month. It provides insights into crime trends by categorizing offenses under major and minor heads, helping to understand the nature and frequency of crimes.

Each entry in the dataset represents a specific type of crime, along with details such as how many cases occurred this month, last month, and the same month last year. This allows for a comparative analysis of crime patterns over time.

The major heads classify crimes into broader categories like Murder, Theft, and Assault, while the minor heads give further details, such as the reason behind a crime (e.g., Murder due to property disputes).

This dataset is useful for tracking crime trends, identifying high-risk crime categories, and helping authorities make informed decisions for public safety and law enforcement improvements. 

# Overview of the Dataset

Total Records: 692
Columns: 8

Data Types:
Numerical: 4 columns (Crime counts, Serial Number)
Categorical: 3 columns (Crime categories and descriptions)
Float: 1 column (Crime counts for the previous year)

Key Columns:
Heads of Crime: Broad classification (e.g., IPC crimes).
MAJOR HEAD: Specific crime category (e.g., Murder).
MINOR HEAD: Sub-categorization of the crime (e.g., Personal Vendetta).
                 
Crime Statistics:
Current Year (Upto Review Month), 
Previous Year (Same Month), 
Previous Month, 
Current Month. 

This dataset tracks crime trends across different categories, allowing for year-over-year and month-over-month comparisons. Let me know if you need further analysis.

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df.index, y=df["During the current month"], label="Current Month")
sns.lineplot(x=df.index, y=df["During the previous month"], label="Previous Month")
plt.xlabel("Crime Index")
plt.ylabel("Crime Count")
plt.title("Crime Trend Over Time")
plt.legend()
plt.show()
```

# Conclusion
The dataset provides a clear picture of crime trends over time by comparing the number of cases in the current month, the previous month, and the same month last year.
It helps in understanding whether crime rates are increasing or decreasing and identifies the most frequently occurring crimes.

The classification of crimes into major and minor heads allows for a deeper understanding of the causes behind each crime, such as personal conflicts, financial disputes, or revenge. This information can be useful for law enforcement agencies to take preventive measures and focus on high-priority crime categories.
While the dataset is useful for tracking crime trends, it lacks location-based information, which could help in identifying crime-prone areas. Additional details like the time of crime, age group of offenders, and police response times would enhance crime analysis further.
